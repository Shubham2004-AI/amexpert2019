```
campaign_dataset = pd.read_csv('drive/My Drive/Projects/customer_recommendation/data/campaign_data.csv')
coupon_item_mapping = pd.read_csv('drive/My Drive/Projects/customer_recommendation/data/coupon_item_mapping.csv')
customer_demographics = pd.read_csv('drive/My Drive/Projects/customer_recommendation/data/customer_demographics.csv')
customer_transaction_data = pd.read_csv('drive/My Drive/Projects/customer_recommendation/data/customer_transaction_data.csv')
item_data = pd.read_csv('drive/My Drive/Projects/customer_recommendation/data/item_data.csv')
train_dataset = pd.read_csv('drive/My Drive/Projects/customer_recommendation/data/train.csv')
```
