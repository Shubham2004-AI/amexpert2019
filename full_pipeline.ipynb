```python
# Importing Libraries
import tensorflow as tf
import pandas as pd
from utility import load_data, merge_data, preprocess_data, X_y, data_vis, any_nan_val
```

```python
# Load the Data
campaign_dataset, coupon_item_mapping, customer_demographics, customer_transaction_data, item_data, train_dataset = load_data()
```

```python
# See all the dataframe 
data_vis(campaign_dataset, coupon_item_mapping, customer_demographics, customer_transaction_data, item_data, train_dataset)
```

```python
# Check if thee is any NaN values
any_nan_val(campaign_dataset, coupon_item_mapping, customer_demographics, customer_transaction_data, item_data, train_dataset)
```

```python
# Merge all the dataset
data = merge_data(campaign_dataset, coupon_item_mapping, customer_demographics, customer_transaction_data, item_data, train_dataset)
```

```python
# Seeing the first 10 rows of dataframe
data.head(10)
```

```python
# Preprocess the Dataframe (to remove the NaN Values)
data = preprocess_data(data)
```

```python
# Seeing the first 10 rows of dataframe
data.head(10)
```

```python
# Making the Model
dense_model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(468, activation='tanh', input_dim=47),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(234, activation='tanh'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(127, activation='tanh'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(64, activation='tanh'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(32, activation='tanh'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(16, activation='tanh'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(8, activation='tanh'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(4, activation='tanh'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(2, activation='tanh'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(1, activation='sigmoid')
    ])
dense_model.summary()
```

```python
# Compiling the Model
dense_model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Making the dataset X and y
X = data.drop('redemption_status',1)
X = pd.get_dummies(X)
y = data.redemption_status
```

```python
# Seeing the X and y dataset
display(X.head(10))

display(y)
```

```python
# Making the training and testing dataframe
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Training the Model
dense_model.fit(X_train, y_train, epochs = 5, batch_size = 100)
```

```python
# Saving the model
dense_model.save('./models/')
```

```python
# Evaluating the model
dense_model.evaluate(X_test, y_test)
```

```python
# Making the predictions
y_pred = dense_model.predict(X_test)
```

```python
# Seeing the y_pred
y_pred
```

```python
# Making y_pred a dataframe
y_pred = pd.DataFrame(y_pred)
```

```python
# Seeing the y_pred
y_pred.head(5)
```

```python
# Making y_pred values in range 0 and 1
for i in y_pred:
  if i > 0.5:
    y_pred[i] = 1
  else:
    y_pred[i] = 0
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
results = confusion_matrix(y_test, y_pred)
```
